a) Use cookiecutter to create the project directory structure. I suggest to use the course cookiecutter template introduced in lecture.

Give the package a name and a short description (see below - we are adding the cipher function). To help you avoid naming collisions on test.pypi.org I suggest to use cipher_<columbia_uni>. E.g. my Columbia Uni is tb2729, so my package would be named cipher_tb2729.
